## Background

I grew up in Wilmington, CA, a small neighborhood in the South Bay and Harbor region of Los Angeles. The area has a predominantly Hispanic/Latino population and is home to a working-class community. Wilmington is located adjacent to the Port of Los Angeles and is home to the third-largest oil field in the continental United States. Several major oil refineries, including Marathon, Phillips 66, Valero, and Tesoro, are based here, making it a critical hub for the regionâ€™s industrial and energy infrastructure.

Living in this environment has deeply inspired me to pursue environmental data science. Growing up surrounded by industrial activity, pollution, and environmental challenges has fueled my desire to contribute something positive to my community. I aim to use my education and skills to work on projects focused on environmental justice, conservation planning, and biodiversity monitoring. Through data-driven solutions, I hope to help create a more sustainable and equitable future for communities like mine.
